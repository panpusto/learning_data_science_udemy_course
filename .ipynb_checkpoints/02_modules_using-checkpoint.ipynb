### define array with np's *arrange* function

```python
months = np.arange(1, 13, 1)
```

```python
months
```

```python
income = 100 + 3 * months
income
```

### months/income chart with green dots

```python
plt.plot(months, income, 'go')
plt.xlabel('month')
plt.ylabel('income')
```

```python
cost = 50 + 10 * months
cost
```

### months/income/cost chart

```python
plt.plot(months, income, 'go', months, cost, 'r^')
```

```python
x = np.arange(-5, 5, 0.1)
plt.plot(x, pow(2, x))
plt.xlabel('Arguments')
plt.ylabel('Values')
```
