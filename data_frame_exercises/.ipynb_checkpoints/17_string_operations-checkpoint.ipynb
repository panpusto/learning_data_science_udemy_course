```python
import pandas as pd
import numpy as np
```

```python
fortune = pd.read_csv('../../course-files/Fortune_500_2017.csv', 
                      usecols=[
                          'Rank',
                          'Title',
                          'Industry',
                          'Hqlocation',
                          'Sector'
                      ])
fortune.head()
```

## make uppercase data in a single column

```python
fortune['Sector'] = fortune['Sector'].str.upper()
```

```python
fortune.head()
```

## search data in column

```python
fortune[fortune['Industry'].str.lower().str.contains('comp')].head()
```

## create new columns with split data from existing column

```python
fortune[['City', 'State']] = fortune['Hqlocation'].str.split(',', expand=True)
```

```python
fortune.head()
```

## Function that builds shortcut from str name 

```python
def build_shortcut(row):
    industry = row['Industry']
    result = ''
    for word in industry.split(' '):
        result += word[0]
    return result
```

### test

```python
build_shortcut({'Industry':'Factory Under Newspaper'})
```

```python
fortune['IndustryShort'] = fortune.apply(build_shortcut, axis=1)
```
