Comment: when a dataset contains a lot of columns, it's better to choose only that columns what we need to our work. It consumes less memory. When you don't choose index (by 'index_col' parameter), it will be auto-generated like above.

```python
fuel = pd.read_csv('../../data_sets/fuel.csv', 
                   usecols=[
                       'vehicle_id',
                       'year',
                       'make',
                       'model',
                       'class',
                       'fuel_type',
                       'combined_mpg_ft1'],
                  index_col='vehicle_id')
fuel.head()
```
