```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('../../data_sets/fuel.csv', 
                   usecols=[
                       'vehicle_id',
                       'year',
                       'make',
                       'model',
                       'class',
                       'fuel_type',
                       'combined_mpg_ft1',
                   ],
                   index_col='vehicle_id',
                   low_memory=False)
fuel.head()
```

### adding NaN values to data for practise purposes

```python
fuel.loc[27705, 'class'] = np.NaN
fuel.loc[26561, 'class'] = np.NaN
fuel.loc[27550, 'class'] = np.NaN
```

```python
fuel.head()
```

### cleansing data with dropna function

```python
fuel.dropna()
```

### cleansing data in columns with at least one NaN value

```python
fuel.dropna(axis='columns')
```

```python
fuel.dropna(axis=1)
```

### cleansing data in rows in columns named 'class' with at least one NaN value

```python
fuel.dropna(subset='class')
```

```python
fuel.head()
```

### cleansing data with changing original variable (with *inplace* parameter)

```python
fuel.dropna(subset=['class'], inplace=True)
```

```python
fuel.head()
```
