```python
import pandas as pd
import numpy as np
```

```python
fuel = pd.read_csv('../../data_sets/fuel.csv', 
                   usecols=[
                       'vehicle_id',
                       'year',
                       'make',
                       'model',
                       'class',
                       'fuel_type',
                       'combined_mpg_ft1',
                   ], 
                   index_col='vehicle_id', 
                   low_memory=False)
fuel.head()
```

```python
fuel.loc[27681, 'combined_mpg_ft1'] = np.NaN
```

```python
fuel.head()
```

### sorting ascending by specific columns

```python
fuel.sort_values(by=['make', 'model']).head()
```

### sorting descending by specific columns

```python
fuel.sort_values(by=['combined_mpg_ft1'], ascending=False).head()
```

### sorting specific columns by descending and ascending

```python
fuel.sort_values(by=['make', 'combined_mpg_ft1'], ascending=[True, False]).head()
```

### NaN value is first

```python
fuel.sort_values(by=['combined_mpg_ft1'], na_position='first').head()
```

```python
fuel.sort_values(by=['combined_mpg_ft1'], na_position='first', inplace=True)
fuel.head()
```

```python
fuel.sort_index().head()
```
