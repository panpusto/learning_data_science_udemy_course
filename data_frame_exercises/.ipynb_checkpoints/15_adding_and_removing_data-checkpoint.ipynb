```python
import pandas as pd
import numpy as np
```

```python
professions = pd.read_csv('../../course-files/Prestige.csv')
professions.head()
```

```python
dict = {'name': 'data_scientist',
        'education': np.NaN, 
        'income': np.NaN, 
        'women': np.NaN, 
        'prestige': np.NaN,
        'census': np.NaN, 
        'type': np.NaN
       }
print(dict)
```

## Adding rows

**append method is depracted** better use pandas.concat()

```python
professions = professions.append(dict, ignore_index=True)
professions.tail()
```

## pandas.concat() method

```python
dict2 = {'name': 'data_scientist2',
        'education': np.NaN, 
        'income': np.NaN,
        'women': np.NaN, 
        'prestige': np.NaN,
        'census': np.NaN, 
        'type': np.NaN
       }
```

```python
professions = pd.concat([professions, pd.DataFrame([dict2])], ignore_index=True)
professions.tail()
```

```python
taxi = professions[professions['name'] == 'taxi.drivers']
taxi
```

## Removing data

 ### removing row

```python
professions.drop(98, inplace=True)
```

```python
professions.tail(10)
```

```python
professions = pd.concat([professions, taxi], ignore_index=True)
professions.tail()
```
