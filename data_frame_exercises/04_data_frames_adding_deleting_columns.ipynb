```python
import pandas as pd
```

```python
fuel = pd.read_csv('../../data_sets/fuel.csv', 
                   usecols=[
                       'vehicle_id',
                       'year',
                       'make',
                       'model',
                       'class',
                       'fuel_type',
                       'combined_mpg_ft1'
                   ], 
                   low_memory=False)
fuel.head()
```

### adding new column (kilometres per liter with values

```python
fuel['combined_kpl'] = fuel['combined_mpg_ft1'] * 1.609 / 3.78
```

```python
fuel.head()
```

```python
fuel.insert(loc=5, column='liters per km', value=100 / fuel['combined_kpl'])
```

```python
fuel.head()
```

### column removal

```python
del fuel['combined_mpg_ft1']
```

```python
fuel.head()
```

```python
fuel.drop(columns='combined_kpl', inplace=True)
```
